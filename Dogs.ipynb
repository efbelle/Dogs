### New dog product
I'm coming out with a new dog product. 
1. Who would be my initial primary targets?
2. Where do they live
3. What type of dogs do they own?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dog1 = pd.read_csv('20151001hundehalter.csv')
dog1['File'] = '2015'
dog1.head()
```

```python
dog2 = pd.read_csv('20160307hundehalter.csv')
dog2['File'] = '2016'
dog2.head()
```

```python
dog3 = pd.read_csv('20170308hundehalter.csv')
dog3['File'] = '2017'
dog3.head()
```

```python
frames = [dog1, dog2, dog3]
dogs= pd.concat(frames)
dogs.head()
```

```python
dogs.columns
```

```python
dogs.rename(columns={'HALTER_ID':'ID', 'ALTER':'Age', 'GESCHLECHT':'Owner Gender', 'STADTKREIS':'District', 'STADTQUARTIER': 'City Quarter',
       'RASSE1': 'Primary Breed',  'RASSE2':'Secondary Breed',  'RASSENTYP':'Breed Type',
       'GEBURTSJAHR_HUND':'Year of birth', 'GESCHLECHT_HUND':'Dog Gender', 'HUNDEFARBE':'Color'}, inplace=True)
```

```python
dogs.head()
```

```python
dogs.drop(['RASSE1_MISCHLING', 'RASSE2_MISCHLING'], axis=1, inplace=True)
```

```python
dogs.head()
```

```python
dogs.drop(['Secondary Breed'], axis=1, inplace=True)
```

```python
dogs.head()
```

```python
dogs.info()
```

```python
dogs['District'].value_counts()
```

```python
districts = dogs['District']
```

```python
owner_gender = dogs['Owner Gender'].value_counts()
```

```python
owner_gender.plot.pie(y='gender', figsize = (5,5))
```

#### There are clearly more women dog owners than men.

```python
dogs['Dog Gender'].value_counts()
```

```python
woman = dogs['Owner Gender'] == 'w'
```

```python
dogs_w = dogs[woman]
dogs_w
```

```python
dogs_w_count = dogs_w['District'].value_counts()
```

```python
#Most popular dog districts
ax = dogs_w_count.plot.bar(x='district', rot =0)
```

#### The top districts in question are 11, 7, 9. We should focus out marketing efforts there

```python
prim_br = dogs_w['Primary Breed'].value_counts()
prim_br1 = prim_br > 200
prim_br2 = prim_br[prim_br1]
```

```python
ax1 = prim_br2.plot.bar()
```

#### It seems that many of these women prefer smaller dogs. If we factor out Labrador Retrievers, most of these dogs are small.

```python
ppl_age = dogs_w['Age'].value_counts()
```

```python
ax2 = ppl_age.plot.bar()
```

#### The women also appear to be older. Many of them above 40.

# Final thoughts

With this new dog toy we are inventing, we'd be better off initially targeting it to women as they make over 70% of the dog ownership population.
These women also appear to be older, mostly above 40. We are dealing with a more mature demographic.
We don't want to be too cutesy. This could also mean that we are dealing with women who are more stable and who have a little big more resources that they can spend.

It also seems that there is a clear bias towards smaller dogs if we don't consider labrador retrievers.
This could be an indication of level of activity.
Smaller dogs can be easily travelled with and so these women may be a bit more on the active side. They could prefer to have smaller dogs that they can easily take with them to various places.

Additionally, we should focus our marketing efforts on women who live in districts 11, 7 and 9. This is where a huge bulk of the women with small dogs live. This doesn't mean we neglect the other districts, just that we initially focus there since this is where a majority of women live.
